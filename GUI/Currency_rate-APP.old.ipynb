```python
import tkinter as tk
from alpha_vantage.foreignexchange import ForeignExchange


window = tk.Tk()

window.title("Currency Exchange Rate")

window.geometry("600x400")

#----Function----
def data_generator():
    
    cc = ForeignExchange(key='3WA8H4TVHIDTGIMP')
    
    firstCurrency = str(entry1.get())
    SecondCurrency = str(entry2.get())
        """ Returns the realtime exchange rate for any pair of digital
        currency (e.g., Bitcoin) or physical currency (e.g., USD).
        Keyword Arguments:
            from_currency: The currency you would like to get the exchange rate
            for. It can either be a physical currency or digital/crypto currency.
            For example: from_currency=USD or from_currency=BTC.
            to_currency: The destination currency for the exchange rate.
            It can either be a physical currency or digital/crypto currency.
            For example: to_currency=USD or to_currency=BTC.
        """
    data, _ = cc.get_currency_exchange_rate(from_currency=firstCurrency,to_currency=SecondCurrency)
    return data


def phrase_display():
    data = data_generator()
    
    #Text creates the text fiedl
    exchangerate_display = tk.Text(master=window, height=5, width=40)
    exchangerate_display.grid(column=1, row=7)
    
    exchangerate_display.insert(tk.END, 'From Currency: ' + data['2. From_Currency Name'] + '\n')
    exchangerate_display.insert(tk.END, 'To Currency:' + data['4. To_Currency Name'] + '\n')
    exchangerate_display.insert(tk.END, 'Exchange Rate: ' + data['5. Exchange Rate'] + '\n')
    exchangerate_display.insert(tk.END, 'Refreshed Time: ' + data['6. Last Refreshed'])


def calculate():
    data = data_generator()
    rate = float(data['5. Exchange Rate'])
    
    firstCurrency_quant = float(entry3.get())
    result_quant = rate * firstCurrency_quant
    
    result_display = tk.Text(master=window, height=6, width=40)
    result_display.grid(column=1, row=8)
    
    result_display.insert(tk.END, str(firstCurrency_quant) + '\n' + data['2. From_Currency Name'] + '\n')
    result_display.insert(tk.END, "=" + '\n')
    result_display.insert(tk.END, str(result_quant) + '\n' + data['4. To_Currency Name'] + '\n')
    result_display.insert(tk.END, 'Refreshed Time: ' + data['6. Last Refreshed'])
    
# Label
label1 = tk.Label(text="Daniel's Currency Exchange!", font=("Times New Roman", 15))
label1.grid(column=0, row=0)

label2 = tk.Label(text="what is input Currency?",font=("Times New Roman", 12))
label2.grid(column=0, row=1)

label2 = tk.Label(text="what is your output Currency?",font=("Times New Roman", 12))
label2.grid(column=0, row=2)

label2 = tk.Label(text="what is the amount of input Currency?",font=("Times New Roman", 12))
label2.grid(column=0, row=3)

#Entry field
entry1=tk.Entry()
entry1.insert(0,'USD')
entry1.grid(column=1, row=1)

entry2=tk.Entry()
entry2.insert(0,'CNY')
entry2.grid(column=1, row=2)

entry3=tk.Entry()
entry3.grid(column=1, row=3)

#Button
button1 = tk.Button(text="show exchange rate",font=("Times New Roman", 12), command = phrase_display)
button1.grid(column=0, row=4)

button2 = tk.Button(text="Calculate",font=("Times New Roman", 12), command = calculate)
button2.grid(column=1, row=4)

#Make the fram appear on screen, Akin to calling a function
window.mainloop()
```
